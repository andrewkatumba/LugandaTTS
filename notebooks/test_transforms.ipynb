```python
import sys
TTS_PATH = "TTS/"
sys.path.append(TTS_PATH)
```

```python
audio_path = '/content/0a4b5961ab945f2a7332dfdaa10fbd5d.wav'
config_path = '/content/config.json'

sample_rate = 22050
encoding = 'PCM_16'

tmp_path = '/content/tmp.wav'
```

Load config:

```python
C = load_config(config_path)
C.audio['do_trim_silence'] = False
ap = AudioProcessor(bits=None, **C.audio)
```

Load audio:

```python
audio = ap.load_wav(audio_path)
wav = np.asarray(audio, dtype=np.float32)
```

Apply mel STFT and inverse mel STFT:

```python
mel = ap.melspectrogram(wav).astype("float32")
out = ap.inv_melspectrogram(mel)
```

Write result to file:

```python
sf.write(tmp_path, out, sample_rate, encoding)
```

Play processed audio:

```python
playsound(tmp_path)
```

Original audio:

```python
playsound(audio_path)
```
